```python
import pandas as pd

# Replace with your actual file name if different
df = pd.read_csv('nvidia.csv')

# Show the first few rows
df.head()
```

```python
# Create the 'Target' column: 1 if next day's close is higher, else 0
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)

# Drop the last row since it has no "next day" to compare
df = df[:-1]

# Check result
df[['Date', 'Close', 'Target']].tail(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Step 1: Select features
features = ['Open', 'High', 'Low', 'Close', 'Volume']
X = df[features]

# Step 2: Target column
y = df['Target']

# Step 3: Scale features to [0, 1]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Optional - check shapes
print("Feature shape:", X_scaled.shape)
print("Target shape:", y.shape)
```

```python
# Define split index (80% for training)
split_idx = int(len(X_scaled) * 0.8)

# Training data
X_train = X_scaled[:split_idx]
y_train = y[:split_idx]

# Testing data
X_test = X_scaled[split_idx:]
y_test = y[split_idx:]

# Print shapes
print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# Evaluate
from sklearn.metrics import classification_report, accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
# We'll work on a fresh copy of the original dataframe
df_feat = df.copy()

# Moving Averages (trend features)
df_feat['MA_5'] = df_feat['Close'].rolling(window=5).mean()
df_feat['MA_10'] = df_feat['Close'].rolling(window=10).mean()

# Price Returns (momentum features)
df_feat['Return_5'] = df_feat['Close'].pct_change(periods=5)
df_feat['Return_10'] = df_feat['Close'].pct_change(periods=10)

# Volatility (risk)
df_feat['Volatility_5'] = df_feat['Close'].rolling(window=5).std()
df_feat['Volatility_10'] = df_feat['Close'].rolling(window=10).std()

# Drop rows with NaN values created by rolling calculations
df_feat = df_feat.dropna().reset_index(drop=True)

# Confirm feature columns
df_feat[['Close', 'MA_5', 'MA_10', 'Return_5', 'Return_10', 'Volatility_5', 'Volatility_10']].head()

```

```python
from sklearn.preprocessing import MinMaxScaler

# Step 1: Select new feature columns
feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume',
                'MA_5', 'MA_10', 'Return_5', 'Return_10', 'Volatility_5', 'Volatility_10']

X = df_feat[feature_cols]
y = df_feat['Target']

# Step 2: Scale features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Train/test split (same 80/20 no-shuffle logic)
split_idx = int(len(X_scaled) * 0.8)
X_train, X_test = X_scaled[:split_idx], X_scaled[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# Confirm shape
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Train new model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict
y_pred_rf = rf.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```
